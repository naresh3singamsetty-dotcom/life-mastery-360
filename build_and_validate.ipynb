```python
import subprocess
from pathlib import Path

workspace = Path(r"c:\Users\nares\life-mastery-360")
backend = workspace / "backend"
frontend = workspace / "frontend"

print('Running backend Maven build...')
result = subprocess.run(['mvn', 'clean', 'install', '-DskipTests'], cwd=backend, capture_output=True, text=True)
print('--- BACKEND EXIT CODE ---')
print(result.returncode)
print(result.stdout)
print(result.stderr)

print('--- FINISHED BACKEND BUILD ---')
print('If backend build succeeds, start the frontend separately with npm install and npm run dev in the frontend folder.')
```

```python
import subprocess

print('Checking for Maven on PATH...')
for cmd in ['mvn', 'mvn.cmd', 'mvn.bat']:
    result = subprocess.run([cmd, '-version'], capture_output=True, text=True)
    print(f'Command: {cmd}')
    print('Return code:', result.returncode)
    print('Stdout:', result.stdout.splitlines()[0] if result.stdout else '')
    print('Stderr:', result.stderr.splitlines()[0] if result.stderr else '')
```

```python
import subprocess

print('Checking for Node and npm...')
for cmd in [['node', '--version'], ['npm', '--version']]:
    try:
        result = subprocess.run(cmd, capture_output=True, text=True)
        print(cmd[0], '->', result.returncode, result.stdout.strip(), result.stderr.strip())
    except FileNotFoundError:
        print(cmd[0], 'not found')
```
